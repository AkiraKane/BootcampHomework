(2) Suppose $x$ is a variable from normal distribution, whose mean is 3 and variance is 25. What is the probability of $-10 < x < 16$.

Make a plot like this:
![normal](http://slides.nycdatascience.com/BOOTCAMP3/content/week6/pic/pro.png)

Replace the "?" with right answer in the plot.

```python
plt.rcParams['figure.figsize'] = 8, 6
xRng = np.arange(-20, 30, .1)
yRng = stats.norm.pdf(xRng, 3, 5)
pRng = np.linspace(-10, 16, 100)

plt.plot(xRng, yRng, color = 'blue')
plt.fill_between(pRng, stats.norm.pdf(pRng, 3, 5), color = 'blue', alpha = 0.5)

plt.xlim(-20, 30)
plt.ylim(0.00, 0.08)

plt.title("Normal distribution: $N(3,25)$")

plt.text(-6.5, 0.01, "$p(-10<x<16)=$ %.3f" % (stats.norm.cdf(16, 3, 5) - stats.norm.cdf(-10, 3, 5)), fontsize=15)
```
